# Data Leaguepedia

```python
import mwclient
import time
import datetime as dt
from datetime import date, timedelta
import json
import ast

# https://lol.fandom.com/wiki/Special:CargoTables/ScoreboardGames
# https://lol.fandom.com/wiki/Special:CargoTables/ScoreboardPlayers

site = mwclient.Site('lol.fandom.com', path='/')
response = site.api('cargoquery',
	limit = "max",
	tables = "ScoreboardGames=SG, ScoreboardPlayers=SP",
	join_on = "SG.GameId=SP.GameId",
	fields = "SG.Tournament, SG.RiotPlatformGameId, SG.MatchId, SG.DateTime_UTC, SG.Team1, SG.Team2, SG.Winner, SG.Patch, SP.Link, SP.Team, SP.Champion, SP.SummonerSpells, SP.Kills, SP.Assists, SP.Deaths, SP.Role, SP.Side, SP.Items, SP.Gold, SP.VisionScore, SP.DamageToChampions",
	where = "SG.Tournament = 'LCS 2022 Summer'"
)

parsed = json.dumps(response)
decoded = json.loads(parsed)

data_leaguepedia = pd.DataFrame(decoded['cargoquery'])

def extraire_variables_imbriquees(df, colonne):
    # Vocabulaire  connaitre : liste/dictionnaire en comprhension
    df[colonne] = [ast.literal_eval(str(item)) for index, item in df[colonne].iteritems()]

    df = pd.concat([df.drop([colonne], axis=1), df[colonne].apply(pd.Series)], axis=1)
    return df


data_leaguepedia = extraire_variables_imbriquees(data_leaguepedia, 'title')

data_leaguepedia.rename(columns={'RiotPlatformGameId' : 'gameid', 'Link' : 'playername'}, inplace=True)

data_leaguepedia
```
